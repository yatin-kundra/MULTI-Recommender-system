# recommendation system

```python
anime = pd.read_csv("anime.csv")
```

```python
anime.head()
```

```python
ratings = pd.read_csv("rating.csv")
```

```python
ratings.head()
```

```python
anime_with_ratings = anime.merge(ratings, on = 'anime_id')
```

```python
anime_with_ratings.drop_duplicates()
```

```python
anime
```

```python
anime_with_ratings.rename(columns={"rating_y": "user_ratings"}, inplace = True)
```

```python
# will be taking only those users who have rated atleast 100 anime
x = anime_with_ratings.groupby('user_id').count()['user_ratings'] > 100
```

```python
anime_lovers = x[x].index
anime_lovers
```

```python
ani_lovers_ratings = anime_with_ratings[anime_with_ratings['user_id'].isin(anime_lovers)]
```

```python
ani_lovers_ratings
```

```python
y = ani_lovers_ratings.groupby("name").count()['user_ratings'] >= 100
```

```python
famous_anime = y[y].index
```

```python
len(famous_anime)   # we got this much anime which have got ratings from ani_lover users and these anime have been rated
# more than 100 times
```

```python
final_ratings = ani_lovers_ratings[ani_lovers_ratings['name'].isin(famous_anime)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index="name", columns='user_id', values = "user_ratings")
```

```python
pt = pt.fillna(0)
```

```python
pt
```

```python
## each book is now a vector in a dimentional space of 4471X24677
# now we will find the cosin similarity and calculate the distance between two vectors i.e. each anime 
# the more closer the anime less the distance between them 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)   # this is the distance btweem 4471 anime with 4471 anime
```

```python
### making a function called recommend we will give name of a book and it will give us recommended books in return
def recommend(anime):
    indx = np.where(pt.index == anime)[0][0]  ## this will give the index of the book in the pt df
    distance = similarity_score[indx]  ## this is the vector of the book that user have input 
    suggestion = sorted(enumerate(distance), key = lambda x:x[1], reverse = True)[1:6]  ## this will tell the distance and index too
    for i in suggestion:
        #print(i[0])    # this give the index 
        print(pt.index[i[0]])
    
```

```python
recommend("Naruto")
```
