```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
pip install nltk
```

```python
def stem(text):
    y = []
    
    for i in text.split():   # splitting
        y.append(ps.stem(i))
    return " ".join(y)   # joing 
```

```python
imdb_final["tags"] = imdb_final["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000, stop_words="english")
```

```python
vectors = cv.fit_transform(imdb_final["tags"]).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()[1:150]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape      # this is the distance of each vecotor with every other vector 
## hence the shape 
similarity = cosine_similarity(vectors)
```

```python
imdb_final['Title'] = imdb2['Series Title']
imdb_final['Tags'] = imdb2['tags']

# imdb_final.drop()
```

```python
imdb_final
```

```python
imdb_final.drop('tags', axis=1)
```

```python
def recommend(movie):
    imdb_final["title_L"] = imdb_final["Title"].apply(lambda x: x.lower())  # making a col. with all the lower case titles
    s_index = imdb_final[imdb_final["title_L"] == movie.lower()].index[0]   # finding the index
    imdb_final.drop("title_L", axis=1, inplace=True)    # removing that temprary lowecase list
    distances = similarity[s_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(imdb_final.iloc[i[0]].Title)
```

```python
recommend("Marvel's Daredevil")
```
