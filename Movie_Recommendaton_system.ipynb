```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

# we want to chage the dictionary into list eg: ["Action", "Adventure", "Fantasy", "Science Fiction"]

# to do this we can create a helper function
```

```python
import ast 
# we are getting this as an string and ast have a function called Literal_eval that converts a string
# into a list.
```

```python
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i["name"])
    return list
```

```python
# this is just the example for the function we have written

convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
# now for keywords
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert_crew(obj):
    list = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            list.append(i["name"])
            count +=1
        else:
            break
    return list
```

```python
movies["cast"] = movies["cast"].apply(convert_crew)
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
# creating a function to fetch director
def fetch_director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            list.append(i["name"])
            break
    return list
    
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
## we need to remove all the sapces form the list eg: christain bale and make it like christainbale 
# so that out model doesnot get confused with christain rogesrs lets say.
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ", "") for i in x])      
# the [] in lambda is called list comprehention function
movies["overview"] = movies["overview"].apply(lambda x:[i.replace(" ", "") for i in x])     
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ", "") for i in x])     
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ", "") for i in x]) 
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ", "") for i in x]) 
```

```python
# its suggested to convert everthing into lowe case so that there may not be any duplicates dur to upprecase letters
df["tag"] = df["tag"].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# stems the words meaning it shortens a word to its root word , we don't want the algoritham to think acting and acted 
# as two diffrent words the main thing is act so PoterStemmer remove the ed, ing, ion etc and give a root word for any
# given word
```

```python
def stem(text):
    y = []
    
    for i in text.split():   # splitting
        y.append(ps.stem(i))
    return " ".join(y)   # joing 
```

```python
df["tag"] = df["tag"].apply(stem)
```

### the working of the Text Vectorization model  :- bag of words

```python
# now we will be converting all the movies to a Vector 
# this is the technique where we will be converting text to vextors and this is called text vectorization
# there are number of techniques to do it,  <<< TRY THEM ON YOUR OWN >>>
# we will be < using bag of words >    
        ## explaination :- here we will be combining all the words , according to our example we
        ## will merge all the tags of all the movies and form a single df and from that we will take
        ## first say 5000 words that are most frequent in this df (new of all the tags for all the movies)
        ## and for reach movie we will make a say table for each movie and how many times did each most 
        ## frequent words out off 5000 was repeated and now each row will be our vecotr
        ## this will be a (5000, 5000) matrix
        ## and we will recommend the movie that will be the closest to the movie that was specified by the
        ## user and then the model will find the say 5 closes vectors of that movies and will return the 
        ## names of that movie, this will be a 5000 dimentinal sapce
        
        ## it is easy to visualize a 5000, 2 matrix as it will be of only 2 dimentinos ,  i.e. 2D
        ## so say we took only 2 most frequent words then a single vector be like a cordinate for a 2D
        ## space (x, y) here (word1, word2) and then we will select the closest vector to a particular movie
        ## the no, of dimentions depends on with how many does the model performs better.
        
```

### Importing the library for the model training

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
# there are alot of hyprer parameters for this class but we will only use few
```

```python
vectors = cv.fit_transform(df["tag"]).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()[1:150]
```

```python
## now we will need ot apply stemming as we can see some words like accident, accident, accident these
## all have the somewhat same meaning and we dont want them, so we will do this on tags using nltk library
## this is natural language processing library
##### this is done before this code block #####

## now we will have to calculate the distance between each vector with every other vector 
## we will not be calculating euclidean distance rather cosin distance this is basically the angle 
## between every vector , we dont use euclidian distance in high dimentionality
## distance is inversely propotion to similarity
```

```python
# importing to claculate the similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape      # this is the distance of each vecotor with every other vector 
## hence the shape 
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# we want to tell the top 5 most similer movies to a particular movie that the user has entered 
# for doing that we need to see the similarity score and to get the max similarity score we need to sort
# but after sorting we will loose the matrix logic that tells how much similer is a particular movie to 
# other movies, this was told by the second index of the matrix [m][n]  like movie m is similarity score 
# with movie n is similarity[m][n]
# so we will use enumerate function to index all the elements
# example:
    ## sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    df["title_L"] = df["title"].apply(lambda x: x.lower())  # making a col. with all the lower case titles
    movie_index = df[df["title_L"] == movie.lower()].index[0]   # finding the index
    df.drop("title_L", axis=1, inplace=True)    # removing that temprary lowecase list
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend("Batman begins")
```
