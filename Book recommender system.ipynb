## popularity based recommendation system
<b> in this we will print the top 50 who have atleast been reated 250 times<b>

```python
# merging the ratings and books data frame
ratings.merge(books, on="ISBN").shape    # there is loss in the total no. as there are raings for some books that are not in 
# books data frame
```

```python
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
ratings_with_name.head()
```

```python
# We want the books with have been rated more than 250 times. 
## 1. we will make groupby on the basis of Book-Title
no_of_ratings = ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
no_of_ratings.rename(columns={"Book-Rating": "No-of-Ratings"}, inplace = True)
```

```python
no_of_ratings.head()   # this data set contains the books and the no. of ratings they have received 
```

```python
ratings_with_name.head()
```

```python
avg_ratings = ratings_with_name.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={"Book-Rating": "No-of-Ratings"}, inplace = True)
```

```python
avg_ratings.head()
```

```python
 # now we will merge no_of_ratings and avg_of_ratings , so that we can take only those who have No-of-Ratings more than 250 
 # and out of those we will take only top 50
popularity = no_of_ratings.merge(avg_ratings, on = "Book-Title")
```

```python
popularity.head()
```

```python
popularity.rename(columns={"No-of-Ratings_x": "No-of-Ratings"}, inplace = True)
```

```python
popularity.rename(columns={"No-of-Ratings_y": "Avg-Ratings"}, inplace = True)
```

```python
popularity[popularity["No-of-Ratings"] > 250]
# only 186 books who have been rated more than 250 times and now we will sort them
popularity = popularity[popularity['No-of-Ratings'] > 250].sort_values("Avg-Ratings", ascending = False).head(50)
```

```python
popularity.head()
```

# Colaborative filtering based

```python
# we will take only those users who have rated more than 200 books meaning only those who read books often and have knowledge of 
# the domain and only those books which have been rated atleast by 50 users by doing this we can make an intellegent system 
# these no. are by trail and error and we are taking these no. cause they gave he best result

x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
padhe_likhe_users
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users) ]
```

```python
y = filtered_ratings.groupby("Book-Title").count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index="Book-Title", columns='User-ID', values = "Book-Rating")
```

```python
final_ratings[final_ratings['Book-Title'] == "The Great Gatsby"]
```

```python
pt = pt.fillna(0)    # here pt is the matrix of users X books where users are those who have rated atleast 200+ books 
# and books that have recieved more than 250 ratings, this tell which book has been rated how much by which user
```

```python
pt
```

```python
## each book is now a vector in a dimentional space of 706X810
# now we will find the cosin similarity and calculate the distance between two vectors i.e. each book 
# the more closer the book less the distance between them 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)   # this is the distance btweem 706 books with 706 books
```

```python
### making a function called recommend we will give name of a book and it will give us recommended books in return
def recommend(book):
    indx = np.where(pt.index == book)[0][0]  ## this will give the index of the book in the pt df
    distance = similarity_score[indx]  ## this is the vector of the book that user have input 
    suggestion = sorted(enumerate(distance), key = lambda x:x[1], reverse = True)[1:6]  ## this will tell the distance and index too
    for i in suggestion:
        #print(i[0])    # this give the index 
        print(pt.index[i[0]])
    
```

```python
rec = recommend("You Belong To Me")
rec
```
